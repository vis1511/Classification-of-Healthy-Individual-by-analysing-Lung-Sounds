```python
#again reading lines from the diagnosis file
lines = open("D:\ICBHI_final_database\ICBHI_final_database\ICBHI_Challenge_diagnosis.txt","r").readlines()
broken = []

for line in lines:
    line = line[:-1]
    broken.append(line.split('\t'))
    
broken[0:4]
```

```python
#creating the is_healthy feature
data["is_healthy"]=0
for i in range(len(data)):
    if data["disease"][i]=="Healthy":
        data["is_healthy"][i]=1
data
```

```python
data.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
#encoding the gender class
for i in range(len(data)):
    if data["Gender"][i]=="F":
        data["Gender"][i]=0
    else:
        data["Gender"][i]=1

#modifying the BMI column to account for NA values
import math
math.isnan(data["BMI"][0])
data["Height"][125]=103
for i in range(len(data)):
    if math.isnan(data["BMI"][i]):
        if (math.isnan(data["Weight"][i]) and math.isnan(data["Height"][i])):
            data["BMI"][i]=22
        else:
            data["BMI"][i]=data["Weight"][i]/(0.0001*data["Height"][i]**2)

```

```python
#dropping the weight and height column as they are no longer required
data.drop(["Weight","Height"],axis=1,inplace=True)
```

```python
data["BMI"].fillna(22,inplace=True)
data["Age"].fillna(20,inplace=True)
data.info()
```

```python
data
```

```python
#saving to a csv file
data.to_csv("demographic_info.csv")
```

Now we will split the data in train-test files

```python
data = pd.read_csv("D:\ICBHI_final_database\ICBHI_final_database\demographic_info.csv")
data.head()
```

```python
data.set_index("Patient_ID",inplace=True)
data.head()
```

```python
#reding lines and seperating them from train test split file
lines = open("D:\ICBHI_final_database\ICBHI_final_database\ICBHI_challenge_train_test.txt","r").readlines()
broken = []

for line in lines:
    line = line[:-1]
    broken.append(line.split('\t'))

broken[0:4]  
```

```python
df=pd.DataFrame(broken,columns=["filename","split"])
df
```

```python
df[df["split"]=='test'].count()
```

```python
df[df["split"]=='train'].count()
```

```python
#create features 
age=[]
gender=[]
bmi=[]
is_healthy=[]
disease=[]
for i in range(len(df)):
    age.append(data["Age"][int(df["filename"][i][0:3])])
    gender.append(data["Gender"][int(df["filename"][i][0:3])])
    bmi.append(data["BMI"][int(df["filename"][i][0:3])])
    is_healthy.append(data["is_healthy"][int(df["filename"][i][0:3])])
    disease.append(data["disease"][int(df["filename"][i][0:3])])
```

```python
#add the created features
df["Age"]=age
df["Gender"]=gender
df["BMI"]=bmi
df["disease"]=disease
df["is_healthy"]=is_healthy

df
```

```python
train = df[df["split"]=="train"]
test = df[df["split"]=="test"]
```
